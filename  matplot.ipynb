```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Pie Chart

```python
areas = ['Markerting','Sales','Development','HR','Customer Support']
budget =[2.5,4,10,1,6]
```

```python
plt.pie(budget,labels=areas, autopct="%0.2f%%")
plt.show()
```

## Scatter Plot

```python
x = [23,24,45,56,78,90,88,34]
y = [43,54,65,76,87,98,90,23]

plt.scatter(x,y)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age vs Salary')
```

# Line Plot

```python
product_id = [1,2,3,4,5,6,7,8]
online_price = [233,456,770,120,222,444,200,300]
offline_price =[400,300,100,333,444,565,565,899]
chor_bazar = [100,200,300,400,150,250,450,500]
```

```python
plt.plot(product_id,online_price,color='red',marker='D',markersize=10)
plt.xlabel('Product')
plt.ylabel('Prices')
plt.title('Product Price Comparision')
plt.show()
```

```python
plt.plot(product_id,online_price,color='red',marker='D',markersize=10,label='Online Prices')
plt.plot(product_id,offline_price,color='yellow',marker='*',markersize=10, label = 'Offfline prices')
plt.plot(product_id,chor_bazar,color='green',marker='+',markersize=10,label='Chor_bazar')
plt.xlabel('Product')
plt.ylabel('Prices')
plt.title('Product Price Comparision')
plt.legend()
plt.show()
```

# Bar Plot

```python
year =[2012,2013,2014,2015,2016,2017,2018]
price =[65000,69000,75000,55000,52000,64000,81000]

plt.bar(year,price,color='red')
plt.xlabel('Year')
plt.ylabel('Prices (in INR)')
plt.title('Year by Year Price comp.')

plt.show()
```

```python
year =[2012,2013,2014,2015,2016,2017,2018]
price =[65000,69000,75000,55000,52000,64000,81000]

plt.barh(year,price,color='red')
plt.xlabel('Year')
plt.ylabel('Prices (in INR)')
plt.title('Year by Year Price comp.')

plt.show()
```
